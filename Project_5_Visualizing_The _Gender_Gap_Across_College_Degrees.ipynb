## Guided Project: Visualizing The Gender Gap In College Degrees

The projects is to explore and visualize gender gap across college degrees. To generate line charts to compare across all degree categories.

```python
#import the libraries
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
women_degrees.head(3)
```

```python
women_degrees.shape
```

There is 42 rows ans 18 columns in the data set. There is one column with the year and other columns are data of the percentage of women enorolled for each degree category.

we will use a layout of 6 rows by 3 columns subplots to generate the line charts. We will group the degrees in to STEM, Liberal arts and other category.

```python
stem_cats = ['Psychology','Biology','Math and Statistics',
             'Physical Sciences','Computer Science','Engineering']

lib_arts_cats = ['Foreign Languages', 'English','Communications and Journalism','Art and Performance', 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 'Education', 
              'Agriculture','Business', 'Architecture'] 
```

### Comparing across Degree categories

A function named rules is created to set the limits and to configure other rules for the chart. Subplots of degree catergories are generated for female and male students.

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

def rules(rules):
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom = "off")
    
```

```python
fig = plt.figure(figsize = (16,20))
n = 1
m = 2
k = 3
z = 0
p = 0
for sp in range(0,17):
    #generate the first column of line charts, STEM degrees
    if sp < 6:
        ax = fig.add_subplot(6,3,n)
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
        rules(rules)
        ax.set_title(stem_cats[sp])
        if n == 16:
            ax.tick_params(labelbottom = "on")
            ax.text(2005, 90, 'Men')
            ax.text(2001, 5, 'Women')
        elif n == 1:
            ax.text(2001, 87, 'Women')
            ax.text(2005, 8, 'Men')            
        n+=3
    #generate the second column of line charts, liberal arts   
    elif (6 <= sp < 11):
        ax = fig.add_subplot(6,3,m)
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[z]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[z]], c=cb_orange, label='Men', linewidth=3)
        rules(rules)
        ax.set_title(lib_arts_cats[z])
        if m == 14:
            ax.tick_params(labelbottom = "on")
            ax.text(2003,40,"Women")
            ax.text(2005,55, "Men")
        elif m == 2:
            ax.text(2001,75,"Women")
            ax.text(2005,20,"Men")
        
        m+=3
        z+=1
    #generate the third column of line charts, other category    
    else:
        ax = fig.add_subplot(6,3,k)
        ax.plot(women_degrees['Year'], women_degrees[other_cats[p]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[p]], c=cb_orange, label='Men', linewidth=3)
        rules(rules)
        ax.set_title(other_cats[p])
        if k == 18:
            ax.tick_params(labelbottom = "on")
            ax.text(2001,30,"Women")
            ax.text(2005,65, "Men")
            ax.tick_params(labelbottom = "on")
        elif k == 3:
            ax.text(2005,5,"Men")
            ax.text(2001,90,"Women")
        k+=3
        p+=1
plt.show()        
```
